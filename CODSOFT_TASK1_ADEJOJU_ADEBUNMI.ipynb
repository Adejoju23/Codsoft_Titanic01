```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**DATA PREPARATION AND PREPROCESSING**

```python
data = pd.read_csv('/content/drive/MyDrive/Data/tested.csv')
data
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#convert object data type to integer
```

```python
new_data = data.drop(['PassengerId','Name','Ticket','Cabin','Fare'],axis=1)
new_data
```

```python
new_data.isnull().sum()
```

```python
new_data
```

```python
new_data.isnull().sum()
```

```python
#filling up the age with their mean
new_data['Age'] = new_data['Age'].fillna(new_data['Age'].mean())
```

```python
new_data.isnull().sum()
```

```python
new_data.info()
```

**ENCODING CATEGORICAL VARIABLES**

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder

le = LabelEncoder()
```

```python
#converting object data type to numerical

new_data['Sex'] = le.fit_transform(new_data['Sex'])
```

```python
new_data.info()
```

```python
new_data['Embarked'] = le.fit_transform(new_data['Embarked'])
```

```python
new_data.info()
```

**DATA VISUALIZATION**

```python
#visualizing survival count

sns.countplot(x='Survived', data=new_data)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# visualising age distribution

sns.histplot(x='Age', data=new_data, bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#visualizing survival by gender

sns.countplot(x='Survived', hue='Sex', data=new_data)
plt.title('Survival by Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.show()
```

```python
#visualizing survival by PClass

sns.countplot(x='Survived', hue='Sex', data=new_data)
plt.title('Survival by Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.show()
```

```python
#visualising correlation matrix

correlation_matrix = new_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#visualizing Embarked Distribution

sns.countplot(x='Embarked', data=new_data)
plt.title('Embarked Distribution')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()
```

**DATA SPLITTING**

```python
y = new_data['Survived']
x = new_data.drop('Survived', axis=1)
```

```python
y
```

```python
x
```

```python
###train_test_split

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

**MODEL TRAINING AND EVALUATION**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(x_train, y_train)
```

```python
predt = logreg.predict(x_test)
predt
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, predt)*100
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(x_train, y_train)
```

```python
predt1 = rfc.predict(x_test)
predt1
```

```python
accuracy_score(y_test,predt1)*100
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(x_train,y_train)
```

```python
predt2 = gbc.predict(x_test)
accuracy_score(y_test, predt2)*100
```

Using LogisticRegression,RandomForestClassifier and GradientBoostingClassifier, the accuracy scores predicted were all 100 percent.
